```python
import pandas as pd
```

```python
ratings = pd.read_csv("ml-100k/u.data", delimiter='\t', names=['user_id','item_id', 'rating' ,'timestamp'], header=None)
ratings
```

```python
with open ('ml-100k/u.item', 'r', encoding="ISO-8859-1") as read_file:
    movies_df = pd.DataFrame(columns = ['ID','movie_name','release_date'])
    for line in read_file:
        values = line.split('|')
        movie_id, movie_name, release_date = values[0], values[1], values[2]
        movie_name = movie_name[0:len(movie_name) - 6]
        # print(movie_id, movie_name, release_date)
        data = [int(movie_id), str(movie_name), release_date]
        df = pd.DataFrame(data=[data] , columns=['ID','movie_name','release_date'])
        movies_df = pd.concat([movies_df, df], ignore_index=True)

    # print(movies_df)
    movies_df.sort_values(by='ID', ascending=True, inplace=True)
read_file.close()

ratings.to_csv('ratings.csv', index=False)
movies_df.to_csv('movies.csv', index=False)
```
